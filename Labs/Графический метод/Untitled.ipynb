```python
import numpy as np
from numpy import linalg as LA
A = np.array([[-1,0,0,0,3],[0,2,4,0,0],[0,1,0,3,2]])
b = np.array([-1,24,6])
d_min = np.array([1,1,0,-2,-1])
d_max = np.array([6,6,4,3,4])
d1 = [1,6]
d2 = [1,6]
d3 = [0,4]
d4 = [-2,3]
d5 = [-1,4]
```

```python
def residual(A, b, x):
    return np.linalg.norm(np.dot(A, x) - b)
```

```python
combinations = []
for i in d1:
    for j in d2:
        for k in d3:
            for l in d4:
                for n in d5:
                    combinations.append([i,j,k,l,n])

len(combinations)
```

```python
best_x = None
min_residual = float('inf')
for combo in combinations:
    x = np.array(combo)
    current_residual = residual(A, b, x)
    if current_residual < min_residual:
        min_residual = current_residual
        best_x = x

print("Наименьшая невязка:", min_residual)
print("Лучший вектор x:", best_x)
```
